# Select New pull request green bottom

# Under Comapare changes, select project_branch to compare to main branch

# After comparing the changes, click on Create pull request green bottom

# Under the Open a pull request window, write the title and Leave a comment on the pull request

# To finalize the pull request, click on Merge Pull Request after the review

# Then click on Confirm Merge
